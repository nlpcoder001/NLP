### Importing the Required Libraries

```python
import random
import nltk
import syllapy
from nltk.corpus import brown
import matplotlib.pyplot as plt
```

#### Creating a list of words using the Corpus

```python
corpus_words=brown.words()
```

#### Randomly Selecting 100 words from the corpus

```python
random_values=[]
for i in range(100):
    random_values.append(random.randint(0,len(corpus_words)))
```

```python
generated_words=[]
for i in random_values:
    generated_words.append(corpus_words[i])
```

#### Classifying the generated words as Simple or Complex based on their syllable count

```python
def classify(word):
    if syllapy.count(word)<3:
        return f"{word} is a Simple Word"
    else:
        return f"{word} is a Complex Word"
```

```python
classified_words=[classify(word) for word in generated_words]
```

#### Printing the Result

```python
for i in classified_words:
    print(i)
```

```python
word_lenghts=[len(word) for word in generated_words]
syllables= [syllapy.count(word) for word in generated_words]
```

```python
plt.hist(word_lenghts, bins=20, color="pink", edgecolor="black")
plt.xlabel("Word Length")
plt.ylabel("Frequency")
plt.title("Frequency distrubution of Word Lenghts")
```

```python
plt.hist(syllables, bins=20, color="aqua", edgecolor="black")
plt.xlabel("Syllable Count")
plt.ylabel("Frequency")
plt.title("Frequency distrubution of Syllables")
```

```python
classified_values=[]
counter=0
for i in classified_words:
    if "Simple" in i:
        counter+=1
classified_values.append(counter)
classified_values.append(100-counter)

plt.pie(classified_values, explode=(0,0.3), labels=["Simple", "Complex"], colors=["Pink", "Aqua"], autopct='%1.1f%%', shadow=True, startangle=180)
plt.title("Percentage of Classified Words", fontdict={"size": 14})
```
